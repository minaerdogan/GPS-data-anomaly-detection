```
import pandas as pd
from google.colab import drive


drive.mount('/content/drive')


file_path = "/content/drive/MyDrive/Arac_Verisi.xlsx"

df = pd.read_excel(file_path, usecols=['datetime', 'latitude', 'longitude', 'speed', 'milesage'])


print(df.head())

print(df.max())

```

```

df['datetime'] = pd.to_datetime(df['datetime'])


df = df.sort_values(by='datetime')

df['lat_diff'] = df['latitude'].diff().abs()
df['lon_diff'] = df['longitude'].diff().abs()
df['time_diff'] = df['datetime'].diff().dt.total_seconds()
```

taking the difference of latitude and longitude of two consecutive data

```

lat_threshold = 0.025
lon_threshold = 0.025



anomalies = df[(df['lat_diff'] > lat_threshold) | (df['lon_diff'] > lon_threshold)]


print(f"Number of anomalies detected: {len(anomalies)}")


```

to find the gps anomalies, determine a certain threshold

```

import math

def degrees_to_km(degree_threshold, latitude):

    km_per_degree_lat = 111  # Latitude constant
    km_per_degree_lon = 111 * math.cos(math.radians(latitude))  # Adjusted for latitude


    lat_km = degree_threshold * km_per_degree_lat
    lon_km = degree_threshold * km_per_degree_lon
    return lat_km, lon_km

# .. degrees at .. latitude
latitude = 39
threshold = 0.025
lat_km, lon_km = degrees_to_km(threshold, latitude)
print(f"Latitude: {lat_km:.2f} km, Longitude: {lon_km:.2f} km")
```

calculates the difference between the datas in certain latitude degree (the data is around 39 dedrees). this helps us to decide which threshold to determine

```
from datetime import timedelta

#its hard for car to move 2.5 km average in 1 minute but not a big anomaly

low_jumps = df[(df['lat_diff'] > 0.025) | (df['lon_diff'] > 0.025)]

anomaly_indices = low_jumps.index


for idx in anomaly_indices:
    if idx > 0:
        prev_row = df.loc[idx - 1]
        curr_row = df.loc[idx]


        time_diff = curr_row['datetime'] - prev_row['datetime']


        if time_diff <= timedelta(minutes=2):
            print("Anomaly Detected:")
            print("Previous Row:")
            print(prev_row)
            print("Current Row:")
            print(curr_row)
            print("-" * 25)
```

```
from datetime import timedelta

#its impossible for car to move 3 km average in 1 minute but not a big anomaly

mid_jumps = df[(df['lat_diff'] > 0.032) | (df['lon_diff'] > 0.032)]


anomaly_indices = mid_jumps.index


for idx in anomaly_indices:
    if idx > 0:
        prev_row = df.loc[idx - 1]
        curr_row = df.loc[idx]


        time_diff = curr_row['datetime'] - prev_row['datetime']


        if time_diff <= timedelta(minutes=2):
            print("Anomaly Detected:")
            print("Previous Row:")
            print(prev_row)
            print("Current Row:")
            print(curr_row)
            print("-" * 25)
```

```
from datetime import timedelta


high_jumps = df[(df['lat_diff'] > 0.04) | (df['lon_diff'] > 0.04)]


anomaly_indices = high_jumps.index    #a car cannot move around 4.2 km in one minute, big anomaly


for idx in anomaly_indices:
    if idx > 0:
        prev_row = df.loc[idx - 1]
        curr_row = df.loc[idx]


        time_diff = curr_row['datetime'] - prev_row['datetime']


        if time_diff <= timedelta(minutes=1):
            print("Anomaly Detected:")
            print("Previous Row:")
            print(prev_row)
            print("Current Row:")
            print(curr_row)
            print("-" * 25)
```

```
import matplotlib.pyplot as plt   #plotting the anomalies


plt.scatter(df['longitude'], df['latitude'], c='green', label='Path', s=10)


plt.scatter(low_jumps['longitude'], low_jumps['latitude'], c='yellow', label='Low Jumps', s=15)

plt.scatter(mid_jumps['longitude'], mid_jumps['latitude'], c='orange', label='Mid Jumps', s=15)

plt.scatter(high_jumps['longitude'], high_jumps['latitude'], c='red', label='High Jumps', s=15)


plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.title('GPS Path with Anomalies')
plt.show()
```

```
#-------------------------------------------------------------------------
from geopy.distance import geodesic

df['prev_latitude'] = df['latitude'].shift()

df['prev_longitude'] = df['longitude'].shift()      #new dataframe to display the differences betvetween two data

df['distance'] = df.apply(
    lambda row: geodesic(
        (row['latitude'], row['longitude']),
        (row['prev_latitude'], row['prev_longitude'])
    ).km if pd.notnull(row['prev_latitude']) else 0, axis=1
)
df['time_diff'] = df['datetime'].diff().dt.total_seconds()
df['speed'] = df['distance'] / (df['time_diff'] / 3600)
df['lat_diff'] = df['latitude'].diff().abs()
df['lon_diff'] = df['longitude'].diff().abs()


```

calculate the time difference between two consecutive data points in hours

```
from sklearn.ensemble import IsolationForest


features = ['distance', 'speed', 'lat_diff', 'lon_diff'] #features
model = IsolationForest(contamination=0.05, random_state=42)  #

df['anomaly'] = model.fit_predict(df[features]) #predicted anomalies added to dataframe as new column


anomalies = df[df['anomaly'] == -1]  #flag anomalies as -1
print(anomalies)
```

```
print(df[features].head()) #lat and long differences in first row are ofcourse Nan so fill them
print(df[features].isnull().sum())
df[features] = df[features].fillna(0)
```

```
from sklearn.cluster import DBSCAN # another algorithm
import numpy as np


from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features_scaled = scaler.fit_transform(df[features])


dbscan = DBSCAN(eps=0.5, min_samples=5)
df['cluster'] = dbscan.fit_predict(features_scaled)


anomalies = df[df['cluster'] == -1]
print(anomalies)

```

```
print(df['anomaly'].value_counts())
print(df['cluster'].value_counts())
num_isolation_forest_anomalies = (df['anomaly'] == -1).sum()
num_dbscan_anomalies = (df['cluster'] == -1).sum() #dbscan found 107 anomalies and isolation forest found 100
```

```
print(f"Number of anomalies detected by DBSCAN: {num_dbscan_anomalies}")
print(f"Number of anomalies detected by Isolation Forest: {num_isolation_forest_anomalies}")
```

```
common_anomalies = df[(df['anomaly'] == -1) & (df['cluster'] == -1)]
print(f"Number of anomalies detected by both models: {common_anomalies.shape[0]}")
```

```
import matplotlib.pyplot as plt
dbscan_anomalies = df[df['cluster'] == -1]

isolation_anomalies = df[df['anomaly'] == -1]

common_anomalies = df[(df['cluster'] == -1) & (df['anomaly'] == -1)]

plt.scatter(df[df['cluster'] != -1]['longitude'], df[df['cluster'] != -1]['latitude'],
            c='green', label='Normal Data', s=10, alpha=0.5)


plt.scatter(dbscan_anomalies['longitude'], dbscan_anomalies['latitude'],
            c='blue', label='DBSCAN Anomalies', s=20, alpha=0.7)


plt.scatter(isolation_anomalies['longitude'], isolation_anomalies['latitude'],
            c='orange', label='Isolation Forest Anomalies', s=20, alpha=0.7)

plt.scatter(common_anomalies['longitude'], common_anomalies['latitude'],
            c='red', label='Anomalies Detected by Both', s=40, edgecolors='black', alpha=1)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.title("Comparison of Anomalies Detected by DBSCAN and Isolation Forest")
plt.show()
```
